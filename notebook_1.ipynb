## Data cleaning and preparation

### Subtask:
Clean the data by handling missing values, outliers, and inconsistencies. Prepare the data for analysis by converting data types, if necessary.


**Reasoning**:
Convert timestamp columns to datetime objects and handle missing values in df2 by dropping rows.



```python
df1['date'] = pd.to_datetime(df1['date'])
df2['Timestamp IST'] = pd.to_datetime(df2['Timestamp IST'], format='%d-%m-%Y %H:%M')
df2.dropna(inplace=True)
display(df1.info())
display(df2.info())
```

**Reasoning**:
Check for outliers in the numerical columns of both dataframes using box plots.



```python
numerical_cols_df1 = df1.select_dtypes(include=np.number).columns
numerical_cols_df2 = df2.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols_df1):
    plt.subplot(3, len(numerical_cols_df1), i + 1)
    plt.boxplot(df1[col])
    plt.title(col)
plt.tight_layout()
plt.show()

plt.figure(figsize=(20, 10))
for i, col in enumerate(numerical_cols_df2):
    plt.subplot(3, 4, i + 1)
    plt.boxplot(df2[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

**Reasoning**:
Based on the box plots, there are outliers in several numerical columns of df2. Remove outliers using the IQR method for the 'Execution Price', 'Size Tokens', 'Size USD', 'Start Position', 'Closed PnL', 'Order ID', 'Fee', 'Trade ID', and 'Timestamp' columns in df2.



```python
numerical_cols_df2_outliers = ['Execution Price', 'Size Tokens', 'Size USD', 'Start Position', 'Closed PnL', 'Order ID', 'Fee', 'Trade ID', 'Timestamp']

for col in numerical_cols_df2_outliers:
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df2 = df2[(df2[col] >= lower_bound) & (df2[col] <= upper_bound)]

display(df2.info())
```
